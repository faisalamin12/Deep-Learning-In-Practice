```python
import tensorflow as tf
```

```python
import numpy as np
import tensorflow.keras
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
teams = ["islamabad","Quetta", "Peshawar","Multan"]
years = [2016,2017,2018,2019]
teams[0] = 11
teams[1] = 12
teams[2] = 13
teams[3] = 14
print(teams)
```

```python
model = tensorflow.keras.Sequential([tensorflow.keras.layers.Dense(units = 1 , input_shape=[1])])
```

```python
model.compile(activation = "sgd", loss="mean_squared_error")
```

```python
model.fit(years,teams,epochs=400)
```

```python
result = model.predict([2020])
if result >= 11:
    print("Winner of PSL 2020 is: Islamabad")
    
elif result >= 12:
    print("Winner of PSL 2020 is: Quetta")
    
elif result >= 13:
    print("Winner of PSL 2020 is: Peshawar")
    
elif result >= 14:
    print("Winner of PSL 2020 is: Multan")
    
```
